```python
import os
import numpy as np
import pandas as pd
import zipfile 
```

```python
def extract(file):
    with zipfile.ZipFile(os.getcwd() + "/Pictures/" + file, 'r') as zip_ref:
        zip_ref.extractall(os.getcwd()+"/Pictures/")
extract("64.zip")
```

```python
waldo_images = os.listdir(os.getcwd() + "/Pictures/64/Waldo")
notwaldo_images = os.listdir(os.getcwd() + "/Pictures/64/notwaldo")
waldo_images = pd.DataFrame(np.column_stack((waldo_images, np.ones(len(waldo_images), dtype = int))))
notwaldo_images = pd.DataFrame(np.column_stack((notwaldo_images, np.zeros(len(notwaldo_images), dtype = int))))
data = pd.concat([waldo_images,notwaldo_images])
images, labels = data[0], data[1]
```

```python
#Code chunk below from https://stackoverflow.com/questions/35286540/display-an-image-with-python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread(os.getcwd() + "/Pictures/64/waldo/" + data.iloc[16][0])
imgplot = plt.imshow(img)
plt.show()
```
